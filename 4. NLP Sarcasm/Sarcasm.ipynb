Get The Data

```python
url = 'https://github.com/dicodingacademy/assets/raw/main/Simulation/machine_learning/sarcasm.json'
```

```python
urllib.request.urlretrieve(url,'sarcasm.json')
```

```python
sentences = []
labels = []
with open('sarcasm.json','r') as file:
  sarcasm = json.load(file)
```

```python
for item in sarcasm:
  print(item)
  break
```

```python
for item in sarcasm:
  sentences.append(item['headline'])
  labels.append(item['is_sarcastic'])
```

```python
print(sentences[0:5])
```

```python
sentences_arr = np.array(sentences)
labels_arr = np.array(labels)
```

```python
sentences_arr[0:5]
```

Split The Data

```python
split = round(len(labels_arr) * 0.8)
split
```

```python
train_sentence = sentences_arr[:split]
train_label = labels_arr[:split]

test_sentence = sentences_arr[split:]
test_label = labels_arr[split:]
```

```python
len(train_sentence)
```

```python
train_sentence[:5]
```

Tokenizing

```python
vocab_size = 12000
embedding_dim = 16
max_length = 120
trunc_type = 'post'
padding_type = 'post'
oov_tok = "<OOV>"
```

```python
sentence_token = tf.keras.preprocessing.text.Tokenizer(oov_token='<OOV>', num_words=vocab_size)
sentence_token.fit_on_texts(sentences_arr)
```

```python
# change the sentence into tokens
tokenized_train_sentence = sentence_token.texts_to_sequences(train_sentence)
tokenized_test_sentence = sentence_token.texts_to_sequences(test_sentence)
```

```python
# Pad the sentence
padded_train_sentence = tf.keras.preprocessing.sequence.pad_sequences(tokenized_train_sentence,
                                                                      maxlen=max_length,
                                                                      padding=padding_type,
                                                                      truncating=trunc_type)
padded_test_sentence = tf.keras.preprocessing.sequence.pad_sequences(tokenized_test_sentence,
                                                                     maxlen=max_length,
                                                                     padding=padding_type,
                                                                     truncating=trunc_type)
```

```python
for sentence in padded_train_sentence:
  print(sentence)
  break
```

Model Creation

```python
model = tf.keras.Sequential([
  tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim,input_length=1),
  tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
# Conv1D needs 3 dimension array object, we only have 2 Dimension so it cannot be used
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
class MyCallBack(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if logs.get('accuracy') > 0.76 and logs.get('val_accuracy') > 0.76:
      self.model.stop_training = True
```

```python
callback = MyCallBack()
```

```python
model.fit(padded_train_sentence, train_label, validation_data=(padded_test_sentence, test_label), epochs=500, callbacks=[callback])
```
